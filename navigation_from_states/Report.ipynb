# Learning Algorithm
Deep Q Learning was used to solve an environment where an agent tries to collect yellow bananas while avoiding blue ones. 

Features of Deep Q Learning
* Deep refers to how a nerual network was used with 2 hidden layers of 64 nodes each. The input to the neural net was 37 states provided by the unity environment that corresponds to agent velocity and perception. The output of the neural net are 4 possible agent actions. 
* Decay exploration: In order for an agent learn an optimal policy, it needs to be allowed explore. Initially the agent is set to try random actions with a probability of 1 but every time step that is decayed by 0.9955 until it reaches a .1 minimal exploration probability.
* Experience replay: This helps with score stability by training using a random batch of historical experience. A queue of 100,000 was used from which 64 experiences were randomly selected during local network updates. 
* Target and local networks: To further break sample correlations and reduce solve time, agent learning was set to happen every 4 time steps while the local network updated every time step.
* Q Learning refers to a model-free (values of actions are learned) algorithm that allows agents to go from no knowledge of an environment to knowing optimal actions for a given state. This is accomplished by observing environment rewards and applying Q-learning.
